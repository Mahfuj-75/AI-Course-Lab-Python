Exercise-1

```python
mark = int(input("Enter mark (0-100): "))

if mark >= 80:
    print("Grade: A+")
elif mark >= 70:
    print("Grade: A")
elif mark >= 60:
    print("Grade: A-")
elif mark >= 50:
    print("Grade: B")
elif mark >= 40:
    print("Grade: C")
elif mark >= 33:
    print("Grade: D")
else:
    print("Grade: F")
```

Exercise-3

```python
n = int(input("Enter n: "))

a = 0
b = 1

for i in range(n-1):
    c = a + b
    a = b
    b = c

print("Fibonacci:", a)
```

```python
def sum_lists(list1, list2):
    result = []
    for i in range(len(list1)):
        result.append(list1[i] + list2[i])
    return result

# Example00
l1 = [3,4,1]
l2 = [6,7,2,8]

print(sum_lists(l1, l2))
```

```python
def get_name(full_name):
    names=full_name.split()
    first=names[0]
    last=names[-1]
    return first, last
name = input("Enter full name: ")
first, last = get_name(name)

print("First name:",first)
print("Last name:",last)
```
