Uniform Cost Search

```python
from queue import PriorityQueue
def ucs(snode, dnode, GRAPH):
  expanded=[]
  frontier= PriorityQueue()

  node=(0,snode, [snode]) #edge cose, snode
  frontier.put(node)
  i=1
  while not frontier.empty():
    node=frontier.get() #select a node for expansio
    u=node[1]
    if u in expanded:
      continue
    print()
    print(u+ "selected. Cost: "+str(node[0]))
    if u==dnode: #goal test
      print("***Goal Test Satisfied***")
      print()
      return node
    expanded.append(u)
    print('['+str(i)+']: '+u+"expanding...")
    i=i+1
    for v in GRAPH[u].keys(): #expanding
      if v not in expanded:
        g=node[0]+GRAPH[u][v]
        path=node[2]+[v]
        frontier.put((g,v,path))
        print(u+"->"+v+", Path: "+str(path)+",Cost: "+str(g))
    #print(frontier.queue)
print('Failed')

```

```python
ucs('S','G',EX1)
```

Greedy BFS

```python
h1={'S':10, 'A':8,'B':5,'C':5,
    'D':4,'E':2,'F':2,'G':0}
```

```python

from queue import PriorityQueue
def fbfs(snode, dnode, GRAPH,heur):
  expanded=[]
  frontier= PriorityQueue()

  node=(heur[snode],snode,[snode],0) #(h,city,path,path cost)
  frontier.put(node)
  i=1
  while not frontier.empty():
    node=frontier.get() #select a node for expansio
    u=node[1]
    if u in expanded:
      continue
    print()
    print(u+ " selected. Cost: "+str(node[0]))
    if u==dnode: #goal test
      print(node)
      print(expanded)
      return node
    expanded.append(u)
    print('['+str(i)+']: '+u+"expanding...")
    i=i+1
    for v in GRAPH[u].keys(): #expanding
      if v not in expanded:
        g=node[3]+GRAPH[u][v]
        path=node[2]+[v]
        frontier.put((heur[v],v,path,g))
        print(u+"->"+v+", Path: "+str(path)+", h: "+str(heur[v]))
    #print(frontier.queue)
print('Failed')

```

```python
fbfs('S','G',EX1,h1)
```
